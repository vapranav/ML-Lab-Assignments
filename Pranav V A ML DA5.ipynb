**In this exercise, you are required to implement the k-means algorithm and apply it to a real-life data set**

**Import libraries**

```python
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

**Reading and editing the dataframe**

```python
data = pd.read_csv('./place.txt', header=None)
data.columns = ['latitude', 'longitude']
data.head()
```

**Plotting the points**

```python
lo = data['longitude']
la = data['latitude']
X = data.iloc[:,[0,1]].values
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

**We can clearly see there are 3 clusters in the dataset**

**KMeans Clustering with 4 clusters**

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3,init='k-means++', random_state=42)
kmeans = model.fit_predict(X)
for i in range(300):
  print(i,kmeans[i] + 1)
```

**Outputting the labels and IDs to output.txt**

```python
file = open('output.txt','a', encoding='utf-8')
for i in range(300):
    tmp = "{id} {labels}".format(id = i,labels = kmeans[i])
    print(tmp)
    file.write(tmp)
file.close()
```

**Plotting the 3 clusters**

```python
plt.scatter(X[kmeans == 0, 0],X[kmeans == 0,1],s = 10,c = 'red',label="Cluster 1")
plt.scatter(X[kmeans == 1, 0],X[kmeans == 1,1],s = 10,c = 'blue',label="Cluster 2")
plt.scatter(X[kmeans == 2, 0],X[kmeans == 2,1],s = 10,c = 'green',label="Cluster 3")
plt.title("K-means clustering")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.show()
```
